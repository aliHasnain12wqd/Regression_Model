#Use width and hieght to find out Dot per inches

```python
import math
def calculate_dpi(row):
    # Calculate the diagonal resolution using the Pythagorean theorem
    pixel_diagonal = math.sqrt(row['Width']**2 + row['Hieght']**2)

    # Calculate DPI
    dpi = pixel_diagonal / row['Inches']

    return dpi

# Apply the DPI calculation to each row
new_df['DPI'] = new_df.apply(calculate_dpi, axis=1)
```

```python
new_df['Gpu'] = new_df['Gpu'].apply(lambda x: x.split(' ')[0])
new_df
```

```python
for i in new_df['OpSys']:

  if 'Window' in i:
    new_df['OpSys'] = new_df['OpSys'].replace(i,'Window')

  elif 'mac' in i:
     new_df['OpSys'] = new_df['OpSys'].replace(i,'Mac')
  else:
    new_df['OpSys'] = new_df['OpSys'].replace(i,'others')
```

```python
cpu = []

for i in new_df['Cpu']:
  if 'i3' in i:
    cpu.append(" ".join(i.split(' ')[:3]))
  elif 'i5' in i:
    cpu.append(" ".join(i.split(' ')[:3]))
  elif 'i7' in i:
    cpu.append(" ".join(i.split(' ')[:3]))
  elif 'AMD' in i:
    cpu.append('AMD Processor')
  else:
    cpu.append('Others')

new_df['Cpu'] = cpu
new_df
```

SPLITING **DATA**

```python
import numpy as np
X = new_df.drop(['Price'], axis = 1)
Y = np.log(new_df['Price'])
```

#Convert whole data into x_train, x_test, y_train, y_test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

#ONE HOT_ENCODING FOR X_TRAIN

```python
from sklearn.preprocessing import OneHotEncoder

encode = OneHotEncoder(sparse=False)
categorical_Columns = ['Company',	'TypeName',	'Cpu', 'Gpu'	,'OpSys']

encoded = encode.fit_transform(x_train[categorical_Columns])
encoded_df = pd.DataFrame(encoded, columns=encode.get_feature_names_out(categorical_Columns))

x_train = x_train.reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)

x_train = pd.concat([x_train,encoded_df], axis = 1).drop(columns =categorical_Columns)
```

#ONE HOT_ENCODING FOR X_TRAIN

```python
encoded_test = encode.transform(x_test[categorical_Columns])
encoded_test_df = pd.DataFrame(encoded_test, columns=encode.get_feature_names_out(categorical_Columns))

# Reset index to avoid misalignment
x_test = x_test.reset_index(drop=True)
encoded_test_df = encoded_test_df.reset_index(drop=True)

# Concatenate the transformed test data and remove the original categorical columns
x_test = pd.concat([x_test.drop(columns=categorical_Columns), encoded_test_df], axis=1)
```

#Linear Regression Models

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

#Predict Values

```python
predict = model.predict(x_test)
predict
```

 # mean_absolute_error

```python
from sklearn.metrics import mean_absolute_error

Mer = mean_absolute_error(y_test,predict)
print("The Mean square error is:", Mer)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, predict)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")
```

# r2_score

```python
from sklearn.metrics import r2_score

R_Squared   = r2_score(y_test,predict)

print('R_Squared_( R^2 ) = ',R_Squared,'%')
```

#Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100)

rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

print("Predicted values:", y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
